```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def solucion(p_1):
    return -p_1*(1-p_1)*(-2*p_1+1)
def solucion_2(p_1):
    return -p_1*(1-p_1)*(2*p_1-1)
def solucion_3(p_1):
    return (1-p_1)*(-2*p_1+1)+p_1*((2*p_1-1)+(-2*(1-p_1)))
def Euler4(t,y0,f):
    h = t[1]-t[0]
    y = np.zeros_like(t)
    y[0] = y0
    for i in range (1, len(y)):
        k1 = f(y[i-1])
        k2 = f(y[i-1]+0.5*k1*h)
        k3 = f(y[i-1]+0.5*k2*h)
        k4 = f(y[i-1]+k3*h)
        y[i] = y[i-1]+(h/6)*(k1+2*k2+2*k3+k4)
    return y 
```

```python
t = np.linspace(0,15,500)
plt.plot(t,Euler4(t,0,solucion_2), linestyle='dashed',label = "[0,1]]")
plt.plot(t,Euler4(t,0.1,solucion_2), linestyle='dashed',label = "[0.1, 0.9]")
plt.plot(t,Euler4(t,0.2,solucion_2), linestyle='dashed',label = "[0.2, 0.8 ]")
plt.plot(t,Euler4(t,0.3,solucion_2), linestyle='dashed',label = "[0.3, 0.7 ]")
plt.plot(t,Euler4(t,0.4,solucion_2), linestyle='dashed',label = "[0.4, 0.6 ]")
plt.plot(t,Euler4(t,0.5,solucion_2), linestyle='dashed',label = "[0.5, 0.5 ]")
plt.plot(t,Euler4(t,0.6,solucion_2), linestyle='dashed',label = "[0.6, 0.4 ]")
plt.plot(t,Euler4(t,0.7,solucion_2), linestyle='dashed',label = "[0.7, 0.3 ]")
plt.plot(t,Euler4(t,0.8,solucion_2), linestyle='dashed',label = "[0.8, 0.2 ]")
plt.plot(t,Euler4(t,0.9,solucion_2), linestyle='dashed',label = "[0.9, 0.1 ]")
plt.plot(t,Euler4(t,1,solucion_2), linestyle='dashed',label = "[1, 0 ]")
plt.ylim(-0.05,1.05)
plt.xlabel("Tiempo [s]")
plt.ylabel(r"$p_1$")
plt.legend(fontsize=6)
```

```python
def solucion(p_1, p_2):
    x = p_1*((1-p_1)*(5-4*p_1-8*p_2)-p_2*(4*p_1-3*p_2))
    y = p_2*(p_1*(4*p_1+8*p_2-5)+(1-p_2)*(4*p_1-3*p_2))  
    return x,y
```

```python
def Euler4_(t,x0,y0,z0,f):
    h = t[1]-t[0]
    x = np.zeros_like(t)
    y = np.zeros_like(t)
    z = np.zeros_like(t)
    x[0] = x0
    y[0] = y0
    z[0] = z0
    for i in range (1, len(y)):
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+0.5*k1[0]*h,y[i-1]+0.5*k1[1]*h)
        k3 = f(x[i-1]+0.5*k2[0]*h,y[i-1]+0.5*k2[1]*h)
        k4 = f(x[i-1]+k3[0]*h, y[i-1]+k3[1]*h)
        x[i] = x[i-1]+(h/6)*(k1[0]+2*k2[0]+2*k3[0]+k4[0])
        y[i] = y[i-1]+(h/6)*(k1[1]+2*k2[1]+2*k3[1]+k4[1])
        z[i] = 1-x[i]-y[i]
    return x,y,z


t = np.linspace(0,30,500)
vals = Euler4_(t,1/3,1/3,1/3,solucion)

x_1 = 0.5
y_1 =0.3
z_1 = 1-x_1-y_1
vals_1 = Euler4_(t,x_1,y_1,z_1,solucion)

x_1 = 0.8
y_1 =0.05
z_1 = 1-x_1-y_1

vals_2 = Euler4_(t,x_1,y_1,z_1,solucion)


x_1 = 0.02
y_1 =0.4
z_1 = 1-x_1-y_1

vals_3 = Euler4_(t,x_1,y_1,z_1,solucion)
x_1 = 1
y_1 =0
z_1 = 1-x_1-y_1

vals_4 = Euler4_(t,x_1,y_1,z_1,solucion)

x_1 = 0
y_1 =0
z_1 = 1-x_1-y_1

vals_5 = Euler4_(t,x_1,y_1,z_1,solucion)



x_1 = 0.1
y_1 =0.9
z_1 = 1-x_1-y_1

vals_6 = Euler4_(t,x_1,y_1,z_1,solucion)

plt.plot(t,vals[0], linestyle='dashed',label = "[1/3,1/3,1/3]]")
plt.plot(t,vals_1[0], linestyle='dashed',label = "[0.5,0.3,0.2]]")
plt.plot(t,vals_2[0], linestyle='dashed',label = "[0.8,0.05,0.15]]")
plt.plot(t,vals_3[0], linestyle='dashed',label = "[0.02,0.4,0.58]]")
plt.plot(t,vals_4[0], linestyle='dashed',label = "[1,0,0]]")
plt.plot(t,vals_5[0], linestyle='dashed',label = "[0,0,1]]")
plt.plot(t,vals_6[0], linestyle='dashed',label = "[0.1,0.9,0]]")
plt.legend(fontsize = 7)
plt.xlabel("Tiempo [s]")
plt.ylabel(r"$p_1$")
plt.ylim(-0.05,1.05)
```
