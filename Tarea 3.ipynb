```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Paso 1: Generar x e y con el nuevo rango
x = np.linspace(-4, 4, 100)  # 100 valores linealmente espaciados entre -6 y 6
y = np.linspace(-4, 4, 100)  # Lo mismo para y

# Paso 2: Definir las matrices y realizar la multiplicaciÃ³n matricial
A = np.array([[4, 1],  # Puedes ajustar estos valores segÃºn sea necesario
              [2, 5]])
# Dado que estamos interesados solo en la matriz resultante para graficar, simplificaremos el proceso
X, Y = np.meshgrid(x, y)  # Crear una malla para x y y

# Vamos a calcular el valor Z basado en una operaciÃ³n que represente la multiplicaciÃ³n matricial
# Para simplificar y dado que el cÃ¡lculo de B era incorrecto en el contexto de multiplicaciÃ³n matricial directa, vamos a recalcular Z directamente
# AquÃ­ asumiremos una operaciÃ³n simplificada para obtener un valor Z para graficar
Z = np.zeros(X.shape)  # Inicializar Z con la misma forma que X e Y

for i in range(len(x)):
    for j in range(len(y)):
        vec = np.array([x[i], 1 - x[i]]) @ A @ np.array([y[j], 1 - y[j]]).reshape(-1, 1)
        Z[j, i] = vec  # Nota: Usamos j, i para indexar Z debido a cÃ³mo meshgrid organiza las coordenadas

# Paso 3: Graficar en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar

surf = ax.plot_surface(X, Y, Z, cmap='viridis')
ax.scatter(1/2, 2/3, 3, color = "red", s=20)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
